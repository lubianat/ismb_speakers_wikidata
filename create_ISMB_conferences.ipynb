```python
import pandas as pd
import wikipedia as wp
html = wp.page("Intelligent_Systems_for_Molecular_Biology").html().encode("UTF-8")
try: 
    df = pd.read_html(html)[1]  # Try 2nd table first as most pages contain contents table first
except IndexError:
    df = pd.read_html(html)[0]
```

```python
df
```

```python
df[['city','country']] = df['Location'].str.split(',',expand=True)
```

```python
df = df[["Conference", "city", "country"]]
```

```python
import re
df["Conference"] = [re.sub(r'(.*)\[.*\]+', '\\1', entry) for entry in df["Conference"]]
```

```python
df["year"] = [name.split(" ")[1] for name in df["Conference"]]
```

```python
df.to_csv("preliminary_table.csv")
```
